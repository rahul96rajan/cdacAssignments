### Use Labeled faces dataset  (having dimensions nearly 3000) available in sklearn. Apply PCA to reduce the dimensions of this dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import fetch_lfw_people

dataset = fetch_lfw_people(min_faces_per_person=60)
```

```python
# print(dataset['DESCR'])
```

```python
X = dataset['data']
y = dataset['target']
```

```python
print(X.shape, "\t", y.shape)
```

```python
dataset.images[1].shape
```

```python
plt.imshow(X[1].reshape(62,47),cmap='gray')
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=290, random_state=1)

pca.fit(X)
```

```python
X_pcs = pca.transform(X) # Reduced from 2914 to 290 dimensions. 10% of original
X_inversed = pca.inverse_transform(X_pcs)
```

```python
figure = plt.figure(figsize=(20,8))

for i in range(0,5):
    true_image = X[i].reshape(62,47)
    pca_image = X_inversed[i].reshape(62,47)

    plt.subplot(2, 5, i+1)
    plt.imshow(true_image, cmap='gray')
    
    plt.subplot(2, 5, i+6)
    plt.imshow(pca_image, cmap='gray')

plt.show()

# Top row actual (2914 features)
# Bottom row PCA reduced (290 features)
```
